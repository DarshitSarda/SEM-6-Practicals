```python
def push(stack,n,cost):
    for i in range(n):
        stack.append(i)
        cost = cost + 1
    return cost

def multipop(stack,n,cost):
    for i in range(n):
        k = int(input("Enter the number of elements you want to multipop:"))
        for i in range(min(len(stack),k)):
            stack.pop()
            cost = cost + 1
        print("Stack after multipop:",stack)
        print("Cost after multipop:",cost)
    return cost


stack = []
cost = 0
operations = 0

n = int(input("Enter the number of elements:"))
operations = operations + n
cost = push(stack,n,cost)
print("Number of operations:",operations)
print(f"Cost after {n} operations:",cost)


n = int(input("Enter number of multipop operations you want to perform:"))
operations = operations + n
cost = multipop(stack,n,cost)

amortized_cost = cost/operations
print("Total cost:",cost)
print("Total operations:",operations)
print("Amortized cost:",round(amortized_cost,3))
```

```python
import math

def polar_angle(point):
    x1, y1 = point
    angle = math.atan2(y1, x1)
    return math.degrees(angle)

def turn_calculator(prev, curr, next):
    x1, y1 = prev 
    x2, y2 = curr
    x3, y3 = next
    cross_product = (x2 - x1) * (y3 - y2) - (y2 - y1) * (x3 - x2)
    return cross_product

def graham_scan(points):
    if len(points) < 3:
        return []
    
    sorted_points = sorted(points, key=polar_angle)
    stack = [sorted_points[0], sorted_points[1]]
    
    for i in range(2, len(sorted_points)):
        while len(stack) > 1 and turn_calculator(stack[-2], stack[-1], sorted_points[i]) < 0:
            stack.pop()
        stack.append(sorted_points[i])
    
    collinear_points = []
    for i in range(len(stack)-2,-1,-1):
        if turn_calculator(stack[i],stack[i+1],stack[-1]) == 0:
            collinear_points.append(stack.pop())
    convex_hull = collinear_points + stack
    return convex_hull

input_points = [(0,0), (0,2), (2,2), (1,1), (1,2), (2,0), (1.5,2)]
convex_hull = graham_scan(input_points)
convex_hull = sorted(convex_hull, key=polar_angle)
print("Convex Hull:", convex_hull)
```

```python
import random


def hiring_problem(candidates):
    interview_cost = 5
    hiring_cost = 10
    days_worked = 0
    pay_per_day = 2
    current_best = 0
    hirings = 0
    candidates_interviewed = 0
    cost = 0
    for candidate in candidates:
        if candidate>current_best:
            current_best = candidate
            hiring_new_cost = interview_cost + hiring_cost
            firing_cost = days_worked * pay_per_day
            cost = cost + firing_cost + hiring_new_cost
            hirings += 1
            days_worked = 0
        else:
            days_worked += 1
            cost += interview_cost
        candidates_interviewed += 1
    return cost,candidates_interviewed,hirings



candidates = random.sample(range(1,11),10)
print("Candidates:",candidates)
cost, candidates_interviewed, hirings = hiring_problem(candidates)

print("Number of candidates interviewed:",candidates_interviewed)
print("Number of candidates hired:",hirings)
print("Total cost:",cost)

formula_cost = 5*10 + hirings * 10
print("Formula Cost:",formula_cost)
```

```python
class node:
    def __init__(self,value,left,right,parent,compare):
        self.value = value
        self.left = left
        self.right = right
        self.parent = parent
        self.compare = compare
        
def toggle_compare(variable):
    return variable^1


def insert_node(value):
    global root_node
    current_value = value
    current_parent = root_node
    current_compare = current_parent.compare
    parent_found = False
    while not parent_found:
        if current_value[current_compare]<current_parent.value[current_compare]:
            if current_parent.left == None:
                new_node = node(current_value,None,None,current_parent,toggle_compare(current_compare))
                current_parent.left = new_node
                parent_found = True
            else:
                current_parent = current_parent.left
                current_compare = toggle_compare(current_compare)
        else:
            if current_parent.right == None:
                new_node = node(current_value,None,None,current_parent,toggle_compare(current_compare))
                current_parent.right = new_node
                parent_found = True
            else:
                current_parent = current_parent.right
                current_compare = toggle_compare(current_compare)
    return root_node

def print_tree(node,indent=0,prefix='root'):
    if indent==0:
        print()
    if node is not None:
        print(f"{'' if indent==0 else '|'*indent}" + "--"*indent + f"{'' if indent==0 else ' '}" + f"{prefix}:({node.value[0]},{node.value[1]})")
        print_tree(node.left,indent+1,"left")
        print_tree(node.right,indent+1,"right")


input_values = [[3, 6], [17, 15], [13, 15], [6, 12], [9, 1], [2, 7], [10, 19]]
root_node = node(input_values[0],None,None,None,0)
for i in range(1,len(input_values)):
    root_node = insert_node(input_values[i])
print_tree(root_node)
```

```python
def ford(possible_paths,graph):
    flow = 0
    possible_paths.sort(key=lambda path: min(graph[path[i]][path[i+1]] for i in range(len(path)-1)))
    
    for path in possible_paths:
        min_edge = float('inf')
        
        for i in range(len(path)-1):
            edge_length = graph[path[i]][path[i+1]]
            min_edge = min(min_edge,edge_length)
        
        for i in range(len(path)-1):
            if graph[path[i]][path[i+1]]==0:
                print(f"Path {path} not possible")
                break
            else:
                graph[path[i]][path[i+1]] -= min_edge
        
        flow = flow+min_edge
    return flow
        
def dfs(graph,source,target,path = []):
    path = path + [source]
    if source==target:
        return [path]
    if source not in graph:
        return []
    
    paths = []
    for neighbor in graph[source]:
        if neighbor not in path:
            new_paths = dfs(graph,neighbor,target,path)
            for new_path in new_paths:
                paths.append(new_path)
    return paths


source = '1'
target = '6'
graph = {
    '1' : {'2' : 8, '3' : 10},
    '2': {'4': 2,'5':7},
    '3': {'2': 3,'5':12},
    '4': {'6':10},
    '5':{'4':4,'6':8}
}

possible_paths = dfs(graph,source,target)
print("The possible paths are:",possible_paths)
maxflow = ford(possible_paths,graph)
print("Maxflow:",maxflow)
```

```python
from numpy import random
import time

# Global variables to count comparisons
hoare_comparisons = 0
lomuto_comparisons = 0
standard_comparisons = 0

def partitionHoare(arr, low, high):
    global hoare_comparisons
    pivot_index = random.randint(low, high + 1)
    arr[high], arr[pivot_index] = arr[pivot_index], arr[high]
    pivot = arr[high]
    i = low - 1
    j = high
    while True:
        i += 1
        while arr[i] < pivot:
            hoare_comparisons += 1
            i += 1
        j -= 1
        while arr[j] > pivot:
            hoare_comparisons += 1
            j -= 1
        if i >= j:
            return j
        arr[i], arr[j] = arr[j], arr[i]

def quickSortHoare(arr, low, high):
    if low < high:
        pi = partitionHoare(arr, low, high)
        quickSortHoare(arr, low, pi)
        quickSortHoare(arr, pi + 1, high)

def partitionLomuto(arr, low, high):
    global lomuto_comparisons
    pivot_index = random.randint(low, high + 1)
    arr[high], arr[pivot_index] = arr[pivot_index], arr[high]
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            lomuto_comparisons += 1
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quickSortLomuto(arr, low, high):
    if low < high:
        pi = partitionLomuto(arr, low, high)
        quickSortLomuto(arr, low, pi - 1)
        quickSortLomuto(arr, pi + 1, high)

def partition(arr, low, high):
    global standard_comparisons
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            standard_comparisons += 1
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)

arrH = random.randint(1000, size=(100))
arrL = arrH.copy()
arr = arrH.copy()
n = len(arrH)
print("Initial array is:",arr)
# Reset comparison counts
hoare_comparisons = 0
lomuto_comparisons = 0
standard_comparisons = 0

start = time.time()
quickSortHoare(arrH, 0, n - 1)
end = time.time()
print("Sorted by Hoare:",arrH)


start1 = time.time()
quickSortLomuto(arrL, 0, n - 1)
end1 = time.time()
print("Sorted by Lomuto:",arrL)


start2 = time.time()
quickSort(arr, 0, n - 1)
end2 = time.time()
print("Sorted by normal:",arr)


print(f'Time Hoare: {end - start}')
print(f'Comparisons Hoare: {hoare_comparisons}')
print(f'Time Lomuto: {end1 - start1}')
print(f'Comparisons Lomuto: {lomuto_comparisons}')
print(f'Time Standard: {end2 - start2}')
print(f'Comparisons Standard: {standard_comparisons}')
```

```python
class node:
    def __init__(self,value,colour,left,right,parent):
        self.value = value
        self.colour = colour
        self.left = left
        self.right = right
        self.parent = parent

def change_colour(colour):
    if colour=='R':
        return 'B'
    else:
        return 'R'
    
def recolour_nodes(current_parent):
    global root_node
    if current_parent.parent!=root_node:
        current_parent.parent.colour = change_colour(current_parent.parent.colour)
        
    current_parent.parent.left.colour = change_colour(current_parent.parent.left.colour)
    current_parent.parent.right.colour = change_colour(current_parent.parent.right.colour)
    
def insert_node(value):
    global root_node
    current_parent = root_node
    current_value = value
    parent_found = False
    
    while not parent_found:
        if current_value<current_parent.value:
            if current_parent.left==None:
                new_node = node(current_value,'R',None,None,current_parent)
                current_parent.left = new_node
                parent_found = True
                
                if current_parent.colour == 'R':
                    if current_parent.parent.left.colour == current_parent.parent.right.colour:
                        recolour_nodes(current_parent)
            else:
                current_parent = current_parent.left
        else:
            if current_parent.right==None:
                new_node = node(current_value,'R',None,None,current_parent)
                current_parent.right = new_node
                parent_found = True
                
                if current_parent.colour == 'R':
                    if current_parent.parent.left.colour == current_parent.parent.right.colour:
                        recolour_nodes(current_parent)
            else:
                current_parent = current_parent.right
    return root_node
        

def print_tree(node,indent=0,prefix='root'):
    if node is not None:
        print(f"{'' if indent==0 else '|'*indent}" + "--"*indent + f"{'' if indent==0 else ' '}" + f"{prefix}:{node.value} {node.colour}")
        print_tree(node.left,indent+1,"left")
        print_tree(node.right,indent+1,"right")        
        
input_values = [55, 66, 27, 19, 51, 83, 57, 72]       
root_node = node(input_values[0],'B',None,None,None)
for i in range(1,len(input_values)):
    root_node = insert_node(input_values[i])
print_tree(root_node)
```
